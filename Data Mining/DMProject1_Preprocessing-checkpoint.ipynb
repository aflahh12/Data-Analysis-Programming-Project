# Data Mining Project 1: Data Preparation and Preprocessing

### Importing Data and Assessing Data Quality

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('heart.csv')

print("=== ORIGINAL DATASET OVERVIEW ===")
print(f"Dataset shape: {df.shape}")
print(f"\nDataset info:")
print(df.info())
print(f"\nFirst 5 rows:")
print(df.head())
print(f"\nDataset description:")
print(df.describe())

# Check for missing values
print(f"\n=== MISSING VALUES CHECK ===")
missing_values = df.isnull().sum()
print(missing_values)
print(f"Total missing values: {missing_values.sum()}")

# Check for duplicates
print(f"\n=== DUPLICATE VALUES CHECK ===")
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# Remove duplicates if any
if duplicates > 0:
    df = df.drop_duplicates()
    print(f"Removed {duplicates} duplicate rows. New shape: {df.shape}")
```

```python
# ============================================================================
# 1. DATA CLEANING
# ============================================================================
print("\n" + "="*50)
print("1. DATA CLEANING")
print("="*50)

# Identify outliers using IQR method for numerical columns
def detect_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

# Numerical columns for outlier detection
numerical_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']

print("Outlier detection and handling:")
outlier_info = {}

for col in numerical_cols:
    outliers, lower_bound, upper_bound = detect_outliers_iqr(df, col)
    outlier_info[col] = {
        'count': len(outliers),
        'lower_bound': lower_bound,
        'upper_bound': upper_bound
    }
    print(f"{col}: {len(outliers)} outliers detected (bounds: {lower_bound:.2f} - {upper_bound:.2f})")

# Handle missing values and outliers by replacing with class-wise mean
def replace_with_class_mean(data, column, target_column='HeartDisease'):
    """Replace missing values and outliers with class-wise mean"""
    # Calculate class-wise means
    class_means = data.groupby(target_column)[column].mean()
    
    # Replace outliers with class-wise mean
    outliers, lower_bound, upper_bound = detect_outliers_iqr(data, column)
    for idx in outliers.index:
        target_class = data.loc[idx, target_column]
        data.loc[idx, column] = class_means[target_class]
    
    # Replace missing values with class-wise mean (if any)
    missing_mask = data[column].isnull()
    if missing_mask.any():
        for target_class in data[target_column].unique():
            class_mask = data[target_column] == target_class
            data.loc[missing_mask & class_mask, column] = class_means[target_class]
    
    return data

# Apply class-wise mean replacement for numerical columns
df_cleaned = df.copy()
for col in numerical_cols:
    if col in df_cleaned.columns:
        df_cleaned = replace_with_class_mean(df_cleaned, col)

# Handle special case for Cholesterol (many zeros which might be missing values)
if 'Cholesterol' in df_cleaned.columns:
    # Treat 0 values in Cholesterol as missing
    cholesterol_zero_mask = df_cleaned['Cholesterol'] == 0
    cholesterol_zero_count = cholesterol_zero_mask.sum()
    print(f"\nCholesterol zero values (treated as missing): {cholesterol_zero_count}")
    
    # Replace zeros with class-wise mean
    class_means_chol = df_cleaned[df_cleaned['Cholesterol'] > 0].groupby('HeartDisease')['Cholesterol'].mean()
    for target_class in df_cleaned['HeartDisease'].unique():
        class_mask = df_cleaned['HeartDisease'] == target_class
        df_cleaned.loc[cholesterol_zero_mask & class_mask, 'Cholesterol'] = class_means_chol[target_class]

print("\nData cleaning completed!")
```

```python
# ============================================================================
# 2. DATA TRANSFORMATION (NORMALIZATION)
# ============================================================================
print("\n" + "="*50)
print("2. DATA TRANSFORMATION - NORMALIZATION")
print("="*50)

def min_max_normalize(data, column, new_min=0, new_max=1):
    """Manual Min-Max Normalization"""
    min_val = data[column].min()
    max_val = data[column].max()
    
    # Apply min-max normalization formula
    normalized = new_min + (new_max - new_min) * (data[column] - min_val) / (max_val - min_val)
    return normalized, min_val, max_val

# Apply min-max normalization to numerical columns
df_normalized = df_cleaned.copy()
normalization_params = {}

print("Applying Min-Max Normalization (0-1 range):")
for col in numerical_cols:
    if col in df_normalized.columns:
        original_range = f"[{df_normalized[col].min():.2f}, {df_normalized[col].max():.2f}]"
        df_normalized[col], min_val, max_val = min_max_normalize(df_normalized, col)
        new_range = f"[{df_normalized[col].min():.2f}, {df_normalized[col].max():.2f}]"
        normalization_params[col] = {'min': min_val, 'max': max_val}
        print(f"{col}: {original_range} -> {new_range}")
```

```python
# ============================================================================
# 3. DATA REDUCTION (FEATURE SELECTION)
# ============================================================================
print("\n" + "="*50)
print("3. DATA REDUCTION - FEATURE SELECTION")
print("="*50)

# First, encode categorical variables for correlation analysis
target = 'HeartDisease'
df_for_correlation = df_normalized.copy()

# Identify and encode categorical columns
categorical_columns = []
for col in df_for_correlation.columns:
    if df_for_correlation[col].dtype == 'object':
        categorical_columns.append(col)

print(f"Categorical columns found: {categorical_columns}")

# Encode categorical variables for correlation analysis
for col in categorical_columns:
    if col != target:
        unique_values = df_for_correlation[col].unique()
        print(f"Encoding {col}: {unique_values}")
        
        # Simple label encoding for correlation calculation
        if len(unique_values) == 2:
            # Binary encoding for 2 categories
            df_for_correlation[col] = (df_for_correlation[col] == unique_values[0]).astype(int)
        else:
            # Label encoding for multiple categories
            for i, value in enumerate(unique_values):
                df_for_correlation.loc[df_for_correlation[col] == value, col] = i

# Calculate correlation with target variable
correlations = {}

print("\nCorrelation with HeartDisease target:")
for col in df_for_correlation.columns:
    if col != target:
        try:
            corr = df_for_correlation[col].corr(df_for_correlation[target])
            correlations[col] = abs(corr) if not np.isnan(corr) else 0  # Use absolute correlation
            print(f"{col}: {corr:.3f}")
        except Exception as e:
            print(f"Could not calculate correlation for {col}: {e}")
            correlations[col] = 0

# Select features based on correlation threshold
correlation_threshold = 0.05  # Lowered threshold to keep more features
selected_features = [col for col, corr in correlations.items() if corr > correlation_threshold]
selected_features.append(target)  # Add target variable

print(f"\nSelected features (correlation > {correlation_threshold}):")
for feature in selected_features[:-1]:  # Exclude target from list
    print(f"- {feature} (correlation: {correlations[feature]:.3f})")

# Create reduced dataset using original df_normalized (not encoded version)
df_reduced = df_normalized[selected_features].copy()
print(f"\nDataset shape after feature selection: {df_reduced.shape}")
```

```python
# ============================================================================
# 4. DATA DISCRETIZATION
# ============================================================================
print("\n" + "="*50)
print("4. DATA DISCRETIZATION")
print("="*50)

df_discretized = df_reduced.copy()

# 4.1 Binarization of categorical attributes (if not already binary)
print("4.1 Binarizing Categorical Attributes:")

# Identify all categorical columns
categorical_cols = []
for col in df_discretized.columns:
    if col != target and df_discretized[col].dtype == 'object':
        categorical_cols.append(col)

print(f"Categorical columns to binarize: {categorical_cols}")

# Create dummy variables for all categorical columns
for col in categorical_cols:
    unique_values = df_discretized[col].unique()
    print(f"Creating binary variables for {col}: {unique_values}")
    
    # Create binary column for each unique value
    for value in unique_values:
        # Clean the value name for column naming
        clean_value = str(value).replace(' ', '_').replace('/', '_')
        new_col = f"{col}_{clean_value}"
        df_discretized[new_col] = (df_discretized[col] == value).astype(int)
    
    # Drop original categorical column
    df_discretized = df_discretized.drop(columns=[col])

print(f"Binarization completed. New shape: {df_discretized.shape}")

# 4.2 Binning of numerical attributes
print("\n4.2 Binning Numerical Attributes:")

def equal_width_binning(data, column, n_bins=3):
    """Equal-width binning"""
    min_val = data[column].min()
    max_val = data[column].max()
    width = (max_val - min_val) / n_bins
    
    bins = []
    labels = []
    for i in range(n_bins):
        lower = min_val + i * width
        upper = min_val + (i + 1) * width
        bins.append(lower)
        if i < n_bins - 1:
            labels.append(f"{column}_bin_{i}")
        else:
            labels.append(f"{column}_bin_{i}")
    bins.append(max_val + 0.001)  # Include maximum value
    
    return pd.cut(data[column], bins=bins, labels=labels, include_lowest=True)

def equal_frequency_binning(data, column, n_bins=3):
    """Equal-frequency binning"""
    return pd.qcut(data[column], q=n_bins, labels=[f"{column}_freq_bin_{i}" for i in range(n_bins)], duplicates='drop')

# Apply binning to selected numerical columns
numerical_cols_for_binning = []
for col in df_discretized.columns:
    if col != target and df_discretized[col].dtype in ['float64', 'int64'] and len(df_discretized[col].unique()) > 10:
        numerical_cols_for_binning.append(col)

for col in numerical_cols_for_binning[:3]:  # Limit to first 3 for demonstration
    print(f"\nBinning {col}:")
    
    # Equal-width binning
    try:
        binned_equal_width = equal_width_binning(df_discretized, col, n_bins=3)
        print(f"Equal-width bins distribution:")
        print(binned_equal_width.value_counts().sort_index())
        
        # Create binary columns for each bin
        for bin_label in binned_equal_width.cat.categories:
            df_discretized[f"{col}_EW_{bin_label}"] = (binned_equal_width == bin_label).astype(int)
        
    except Exception as e:
        print(f"Could not perform equal-width binning for {col}: {e}")
    
    # Equal-frequency binning
    try:
        binned_equal_freq = equal_frequency_binning(df_discretized, col, n_bins=3)
        print(f"Equal-frequency bins distribution:")
        print(binned_equal_freq.value_counts().sort_index())
        
        # Create binary columns for each bin
        for bin_label in binned_equal_freq.cat.categories:
            df_discretized[f"{col}_EF_{bin_label}"] = (binned_equal_freq == bin_label).astype(int)
            
    except Exception as e:
        print(f"Could not perform equal-frequency binning for {col}: {e}")
```

```python
# ============================================================================
# FINAL PROCESSING AND OUTPUT
# ============================================================================
print("\n" + "="*50)
print("FINAL PROCESSING AND SUMMARY")
print("="*50)

# Remove original numerical columns that were binned (optional)
# Keep both original and binned for comparison
print(f"Final dataset shape: {df_discretized.shape}")
print(f"Final columns ({len(df_discretized.columns)}):")
for i, col in enumerate(df_discretized.columns, 1):
    print(f"{i:2d}. {col}")

# Check final data types
print(f"\nFinal data types:")
print(df_discretized.dtypes.value_counts())

# Summary statistics
print(f"\nTarget variable distribution:")
print(df_discretized[target].value_counts())

# Save processed dataset
output_filename = 'heart_processed.csv'
df_discretized.to_csv(output_filename, index=False)
print(f"\nProcessed dataset saved as '{output_filename}'")

# Create summary report
print(f"\n" + "="*50)
print("PREPROCESSING SUMMARY REPORT")
print("="*50)
print(f"Original dataset shape: {df.shape}")
print(f"Final dataset shape: {df_discretized.shape}")
print(f"Features removed by correlation threshold: {len(df.columns) - len(selected_features)}")
print(f"Duplicate rows removed: {duplicates}")
print(f"Outliers handled: {sum([info['count'] for info in outlier_info.values()])}")
print(f"Normalization applied to: {len(numerical_cols)} numerical columns")
print(f"Binary encoding applied to categorical variables")
print(f"Binning applied to selected numerical variables")

# Visualization of the preprocessing steps
plt.figure(figsize=(15, 10))
```
