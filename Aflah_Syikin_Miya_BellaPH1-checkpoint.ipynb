## 2.0 DATA PREPARATION AND CLEANING

Steps 1: Prepare both datasets and match the car registration from Cars2025.csv with the information on the carbon emission from FuelConsumption.csv

## 2.1 Importing Datasets

```python
#import necessary library
import pandas as pd

cars_df = pd.read_csv("cars_2025.csv")
fuel_df = pd.read_csv("FuelConsumption.csv")
```

## 2.2 Load Datasets

```python
#show the shape of the cars dataset
print("Cars dataset shape:", cars_df.shape)

#preview the first 5 rows of the cars dataset
cars_df.head()

#show the shape of the fuel consumption dataset
print("Fuel consumption dataset shape:", fuel_df.shape)

#preview the first 5 rows of the fuel consumption dataset
fuel_df.head()
```

Steps 2: Do appropriate data cleaning, preparation and wrangling.

## 2.3 Data Merging and Matching

```python
#clean column names
def clean_column(df):
    return df.rename(columns=lambda x: x.strip().lower().replace(' ', '_'))

#clean for both datasets
cars_df = clean_column(cars_df)
fuel_df = clean_column(fuel_df)

#rename 'maker' to 'make' in cars_df to ensure both datasets have matching column names for merging
cars_df.rename(columns={'maker': 'make'}, inplace=True)

#show the cleaned column names for both datasets
print("\nColumn names after standardization:")
print("\nCars 2025 columns:", cars_df.columns.tolist())
print("\nFuel consumption columns:", fuel_df.columns.tolist())
```

## 2.4 Data Cleaning and Handling Missing Values

```python
#standardize 'make' and 'model' values
for df in [cars_df, fuel_df]:
    df['make'] = df['make'].str.upper().str.strip()
    df['model'] = df['model'].str.upper().str.strip()
    
#remove duplicate rows
cars_before = cars_df.shape[0]
fuel_before = fuel_df.shape[0]

#drop duplicate rows
cars_df.drop_duplicates(inplace=True)
fuel_df.drop_duplicates(inplace=True)

print(f"\nDuplicates removed: Cars - {cars_before - cars_df.shape[0]}, Fuel - {fuel_before - fuel_df.shape[0]}")

#merge both datasets on 'make' and 'model' columns using inner join
merged_df = pd.merge(cars_df, fuel_df, on=['make', 'model'], how='inner')

#show the shape of the merged dataset
print("\nMerged dataset shape:", merged_df.shape)

#check and show the number of missing values in each column before removing them
print("\nMissing values in merged dataset (before):")
print(merged_df.isnull().sum())

#drop rows that containing any missing values from the merged dataset
merged_df.dropna(inplace=True)

#check and show the number of missing values in each column after removing them
print("\nMissing values in merged dataset (after):")
print(merged_df.isnull().sum())

#print the shape of the dataset
print("Shape after remove the missing data:", merged_df.shape)
```

## 2.5 Data Aggregation and Group Operations

```python
# Car models that linked to high CO2 emissions
if all(col in merged_df.columns for col in ['make', 'model', 'fuel', 'co2emissions']):
    high_emitters = merged_df.groupby(['make', 'model', 'fuel'])['co2emissions'].mean().reset_index()
    high_emitters = high_emitters.sort_values(by='co2emissions', ascending=False)
    print("\nTop 5 high CO2 emitting car models (by Make, Model, Fuel Type):")
    print(high_emitters.head())
else:
    print("The required columns for identifying high CO2 emission models not found.")

# average CO2 emissions by fuel type
if 'fuel' in merged_df.columns and 'co2emissions' in merged_df.columns:
    fuel_impact = merged_df.groupby('fuel')['co2emissions'].mean().reset_index()
    fuel_impact = fuel_impact.sort_values(by='co2emissions', ascending=False)
    print("\nAverage CO2 emissions by fuel type:")
    print(fuel_impact)
else:
    print("Columns 'fuel' or 'co2emissisons' not found.")

#hybrid vs overall co2 emissions
if 'fuel' in merged_df.columns and 'co2emissions' in merged_df.columns:
    hybrid_avg = merged_df[merged_df['fuel'].str.contains('HYBRID', case=False, na=False)]['co2emissions'].mean()
    overall_avg = merged_df['co2emissions'].mean()
    reduction = overall_avg - hybrid_avg
    print(f"\nOverall average CO2 emissions: {overall_avg:.2f} g/km")
    print(f"Average CO2 emissions for hybrids: {hybrid_avg:.2f} g/km")
    print(f"Potential CO2 reduction if all cars were hybrid: {reduction:.2f} g/km")
else:
    print("Required columns for hybrid effectiveness is not found.")

#CO2 emissisons by vehicle class
if 'vehicleclass' in merged_df.columns and 'co2emissions' in merged_df.columns:
    class_emissions = merged_df.groupby('vehicleclass')['co2emissions'].mean().reset_index()
    class_emissions = class_emissions.sort_values(by='co2emissions', ascending=False)
    print("\nAverage CO2 emissions by vehicle class:")
    print(class_emissions)
else:
    print("Columns 'vehicleclass' or 'co2emissions' not found for vehicle class.")

#total car registrations by 'make'
if 'make' in merged_df.columns:
    car_counts = merged_df['make'].value_counts().reset_index()
    car_counts.columns = ['make', 'registration_count']
    print("\nTotal car registrations by Make:")
    print(car_counts.head())
else:
    print("Column 'make' not found for car registration count.")

#fuel type distribution
if 'fuel' in merged_df.columns:
    fuel_distribution = merged_df['fuel'].value_counts().reset_index()
    fuel_distribution.columns = ['fuel_type', 'count']
    print("\nFuel type distributions:")
    print(fuel_distribution.head())
else:
    print("Column 'fuel' not found.")
```
